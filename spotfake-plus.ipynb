```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm

# Thư viện AI
import torch
from transformers import XLNetTokenizer, XLNetModel
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import Model

# ==========================================
# 1. CẤU HÌNH ĐƯỜNG DẪN 
# ==========================================
CSV_PATH = '/kaggle/working/data_extracted/AAAI_dataset/politi_train.csv'
IMAGE_DIR = '/kaggle/working/data_extracted/AAAI_dataset/Images/politi_train'
OUTPUT_DIR = '/kaggle/working/' # Nơi lưu file .npy sau khi trích xuất

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"[INFO] Thiết bị xử lý: {device}")

# ==========================================
# 2. KHỞI TẠO MÔ HÌNH (XLNet & VGG19)
# ==========================================
print("[INFO] Đang tải mô hình XLNet...")
tokenizer = XLNetTokenizer.from_pretrained('xlnet-base-cased')
xlnet_model = XLNetModel.from_pretrained('xlnet-base-cased').to(device)
xlnet_model.eval()

print("[INFO] Đang tải mô hình VGG19...")
base_vgg = VGG19(weights='imagenet', include_top=True)
vgg_model = Model(inputs=base_vgg.input, outputs=base_vgg.get_layer('fc1').output)

# ==========================================
# 3. ĐỌC DỮ LIỆU & TRÍCH XUẤT ĐẶC TRƯNG
# ==========================================
df = pd.read_csv(CSV_PATH)
# Lọc bỏ các dòng bị rỗng (NaN) ở cột content hoặc image nếu có
df = df.dropna(subset=['content', 'image']).reset_index(drop=True)

num_samples = len(df)
print(f"[INFO] Tổng số bài báo cần xử lý: {num_samples}")

# Khởi tạo ma trận trống
text_matrix = np.zeros((num_samples, 50, 768), dtype=np.float32)
image_matrix = np.zeros((num_samples, 4096), dtype=np.float32)
labels = np.zeros((num_samples,), dtype=np.int32)

valid_indices = [] # Lưu lại các index xử lý thành công (không bị lỗi ảnh)

print("[INFO] Bắt đầu trích xuất (vui lòng đợi)...\n")
for idx, row in tqdm(df.iterrows(), total=num_samples):
    
    img_filename = str(row['image'])
    img_path = os.path.join(IMAGE_DIR, img_filename)
    
    #  TRÍCH XUẤT ẢNH (VGG19)
    try:
        # Thử mở và đọc ảnh
        img = Image.open(img_path).convert('RGB')
        img = img.resize((224, 224))
        
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)
        
        # Đưa qua VGG19
        img_feature = vgg_model.predict(img_array, verbose=0)
        image_matrix[idx] = img_feature[0]
        
    except Exception as e:
        # Nếu ảnh lỗi hoặc không tồn tại, bỏ qua bài báo này
        continue
        
    #  TRÍCH XUẤT VĂN BẢN (XLNET)
    text_content = str(row['content'])
    inputs = tokenizer(
        text_content, 
        return_tensors="pt", 
        max_length=50, 
        padding='max_length', 
        truncation=True
    ).to(device)
    
    with torch.no_grad():
        outputs = xlnet_model(**inputs)
        # Lấy hidden states, chuyển về CPU để lưu vào numpy
        hidden_states = outputs.last_hidden_state.squeeze(0).cpu().numpy()
        text_matrix[idx] = hidden_states
        
    #  LƯU NHÃN
    labels[idx] = int(row['label'])
    
    valid_indices.append(idx)

# ==========================================
# 4. LƯU KẾT QUẢ
# ==========================================
# Chỉ lấy những bài báo xử lý thành công (lọc bỏ các mảng zero của ảnh lỗi)
final_text_matrix = text_matrix[valid_indices]
final_image_matrix = image_matrix[valid_indices]
final_labels = labels[valid_indices]

print(f"\n[INFO] Hoàn tất! Xử lý thành công {len(valid_indices)}/{num_samples} bài báo.")
print(f"Shape của Text Matrix: {final_text_matrix.shape}")
print(f"Shape của Image Matrix: {final_image_matrix.shape}")
print(f"Shape của Labels: {final_labels.shape}")

np.save(os.path.join(OUTPUT_DIR, 'politi_train_text.npy'), final_text_matrix)
np.save(os.path.join(OUTPUT_DIR, 'politi_train_image.npy'), final_image_matrix)
np.save(os.path.join(OUTPUT_DIR, 'politi_train_labels.npy'), final_labels)

print("\n[INFO] Đã lưu 3 file .npy vào thư mục /kaggle/working/!")
```

```python
import numpy as np
from tensorflow.keras.layers import Dense, Flatten, Dropout, Input, BatchNormalization, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers, initializers, optimizers
from tensorflow.keras.utils import to_categorical

# ==========================================
# 1. TẢI DỮ LIỆU ĐÃ TRÍCH XUẤT
# ==========================================
train_text = np.load('/kaggle/working/politi_train_text.npy')
train_image = np.load('/kaggle/working/politi_train_image.npy')
train_labels = np.load('/kaggle/working/politi_train_labels.npy')

# Chuyển nhãn (0, 1) sang dạng One-Hot (ví dụ: 0 -> [1, 0]; 1 -> [0, 1])
train_labels_cat = to_categorical(train_labels, num_classes=2)

print(f"Kích thước Text: {train_text.shape}")
print(f"Kích thước Image: {train_image.shape}")
print(f"Kích thước Label One-hot: {train_labels_cat.shape}")

# ==========================================
# 2. XÂY DỰNG KIẾN TRÚC SPOTFAKE+
# ==========================================
def build_spotfake_plus():
    # --- NHÁNH TEXT (Đầu vào từ XLNet) ---
    input_text = Input(shape=(50, 768), name="input_text")
    text_flat = Flatten()(input_text)
    
    dense_text = Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(text_flat)
    dense_text = Dense(500, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_text)
    dense_text = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_text)
    
    dense_text = BatchNormalization()(dense_text)
    dense_text_drop = Dropout(0.4)(dense_text)

    # --- NHÁNH IMAGE (Đầu vào từ VGG19) ---
    input_image = Input(shape=(4096,), name="input_image")
    
    dense_image = Dense(2000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(input_image)
    dense_image = Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_image)
    dense_image = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_image)
    
    dense_image = BatchNormalization()(dense_image)
    dense_image_drop = Dropout(0.4)(dense_image)

    # --- GIAI ĐOẠN FUSION & CLASSIFICATION ---
    concat = concatenate([dense_text_drop, dense_image_drop])

    inter_dense = Dense(200, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(concat)
    inter_dense = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(inter_dense)
    final_dense = Dense(50, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(inter_dense)
    
    final_dropout = Dropout(0.4)(final_dense)
    output = Dense(2, activation='softmax', name="output_layer")(final_dropout)

    # Khởi tạo và Compile mô hình
    model = Model(inputs=[input_text, input_image], outputs=output)
    adam = optimizers.Adam(learning_rate=1e-4) # Tốc độ học chuẩn của SpotFake+
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    
    return model

print("\n[INFO] Đang khởi tạo mô hình SpotFake+...")
model = build_spotfake_plus()

# ==========================================
# 3. HUẤN LUYỆN (TRAINING)
# ==========================================
print("\n[INFO] BẮT ĐẦU TRAINING...")
history = model.fit(
    x=[train_text, train_image], 
    y=train_labels_cat,
    batch_size=32,      
    epochs=40,          # Có thể tăng/giảm số epoch tùy ý
    validation_split=0.2 # Trích 20% dữ liệu ra để đánh giá (validation)
)

print("\n[INFO] HOÀN TẤT HUẤN LUYỆN!")
```

```python
import numpy as np
from tensorflow.keras.layers import Dense, Flatten, Dropout, Input, BatchNormalization, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers, initializers, optimizers
from tensorflow.keras.utils import to_categorical

# ==========================================
# 1. TẢI DỮ LIỆU ĐÃ TRÍCH XUẤT
# ==========================================
train_text = np.load('/kaggle/working/gossip_train_text.npy')
train_image = np.load('/kaggle/working/gossip_train_image.npy')
train_labels = np.load('/kaggle/working/gossip_train_labels.npy')

# Chuyển nhãn (0, 1) sang dạng One-Hot (ví dụ: 0 -> [1, 0]; 1 -> [0, 1])
train_labels_cat = to_categorical(train_labels, num_classes=2)

print(f"Kích thước Text: {train_text.shape}")
print(f"Kích thước Image: {train_image.shape}")
print(f"Kích thước Label One-hot: {train_labels_cat.shape}")

# ==========================================
# 2. XÂY DỰNG KIẾN TRÚC SPOTFAKE+
# ==========================================
def build_spotfake_plus():
    # --- NHÁNH TEXT (Đầu vào từ XLNet) ---
    input_text = Input(shape=(50, 768), name="input_text")
    text_flat = Flatten()(input_text)
    
    dense_text = Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(text_flat)
    dense_text = Dense(500, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_text)
    dense_text = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_text)
    
    dense_text = BatchNormalization()(dense_text)
    dense_text_drop = Dropout(0.4)(dense_text)

    # --- NHÁNH IMAGE (Đầu vào từ VGG19) ---
    input_image = Input(shape=(4096,), name="input_image")
    
    dense_image = Dense(2000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(input_image)
    dense_image = Dense(1000, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_image)
    dense_image = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(dense_image)
    
    dense_image = BatchNormalization()(dense_image)
    dense_image_drop = Dropout(0.4)(dense_image)

    # --- GIAI ĐOẠN FUSION & CLASSIFICATION ---
    concat = concatenate([dense_text_drop, dense_image_drop])

    inter_dense = Dense(200, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(concat)
    inter_dense = Dense(100, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(inter_dense)
    final_dense = Dense(50, activation='relu', kernel_regularizer=regularizers.l2(0.01), kernel_initializer=initializers.he_normal(seed=0))(inter_dense)
    
    final_dropout = Dropout(0.4)(final_dense)
    output = Dense(2, activation='softmax', name="output_layer")(final_dropout)

    # Khởi tạo và Compile mô hình
    model = Model(inputs=[input_text, input_image], outputs=output)
    adam = optimizers.Adam(learning_rate=1e-4) # Tốc độ học chuẩn của SpotFake+
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    
    return model

print("\n[INFO] Đang khởi tạo mô hình SpotFake+...")
model = build_spotfake_plus()

# ==========================================
# 3. HUẤN LUYỆN (TRAINING)
# ==========================================
print("\n[INFO] BẮT ĐẦU TRAINING...")
history = model.fit(
    x=[train_text, train_image], 
    y=train_labels_cat,
    batch_size=32,      
    epochs=40,          # Có thể tăng/giảm số epoch tùy ý
    validation_split=0.2 # Trích 20% dữ liệu ra để đánh giá (validation)
)

print("\n[INFO] HOÀN TẤT HUẤN LUYỆN!")
```

```python
import numpy as np
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ==========================================
# XỬ LÝ MẤT CÂN BẰNG DỮ LIỆU (CLASS WEIGHT)
# ==========================================
y_integers = np.argmax(train_labels_cat, axis=1)

# Máy tính sẽ tự động tính xem nhãn nào ít xuất hiện thì phạt nặng hơn
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_integers),
    y=y_integers
)
class_weight_dict = dict(enumerate(class_weights))
print(f"[INFO] Trọng số phạt nhãn: {class_weight_dict}")

# ==========================================
# EarlyStopping: Chống học vẹt (Overfitting)
early_stopping = EarlyStopping(
    monitor='val_loss',        
    patience=6,               
    restore_best_weights=True, # Tự động khôi phục lại bộ trọng số ở vòng có kết quả
    verbose=1
)

#  ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,                
    patience=3,                
    min_lr=1e-6,             
    verbose=1
)

model_checkpoint = ModelCheckpoint(
    filepath='/kaggle/working/spotfake_plus_gossipcop_best.h5',
    monitor='val_accuracy',
    save_best_only=True,     
    verbose=1
)

# ==========================================
# BẮT ĐẦU HUẤN LUYỆN 
# ==========================================
history = model.fit(
    x=[train_text, train_image], 
    y=train_labels_cat,
    batch_size=32,      
    epochs=100, #
    validation_split=0.2,
    class_weight=class_weight_dict, # Ép mô hình chú ý vào Tin giả (Fake)
    callbacks=[early_stopping, reduce_lr, model_checkpoint] 
)

```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
import torch
from transformers import XLNetTokenizer, XLNetModel
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical

# ==========================================
# 1. CẤU HÌNH ĐƯỜNG DẪN TẬP TEST
# ==========================================
TEST_CSV_PATH = '/kaggle/working/data_extracted/AAAI_dataset/gossip_test.csv'
TEST_IMAGE_DIR = '/kaggle/working/data_extracted/AAAI_dataset/Images/gossip_test'

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ==========================================
# 2. TRÍCH XUẤT ĐẶC TRƯNG TẬP TEST
# ==========================================
tokenizer = XLNetTokenizer.from_pretrained('xlnet-base-cased')
xlnet_model = XLNetModel.from_pretrained('xlnet-base-cased').to(device)
xlnet_model.eval()

base_vgg = VGG19(weights='imagenet', include_top=True)
vgg_model = Model(inputs=base_vgg.input, outputs=base_vgg.get_layer('fc1').output)

df_test = pd.read_csv(TEST_CSV_PATH).dropna(subset=['content', 'image']).reset_index(drop=True)
num_samples = len(df_test)

test_text_matrix = np.zeros((num_samples, 50, 768), dtype=np.float32)
test_image_matrix = np.zeros((num_samples, 4096), dtype=np.float32)
test_labels = np.zeros((num_samples,), dtype=np.int32)
valid_indices = []

print(f"\n[INFO] Đang test trên {num_samples} bài báo của tập Test...")
for idx, row in tqdm(df_test.iterrows(), total=num_samples):
    img_path = os.path.join(TEST_IMAGE_DIR, str(row['image']))
    
    # Rút trích Ảnh
    try:
        img = Image.open(img_path).convert('RGB').resize((224, 224))
        img_array = preprocess_input(np.expand_dims(img_to_array(img), axis=0))
        test_image_matrix[idx] = vgg_model.predict(img_array, verbose=0)[0]
    except Exception:
        continue 
        
    # Rút trích Chữ
    inputs = tokenizer(str(row['content']), return_tensors="pt", max_length=50, padding='max_length', truncation=True).to(device)
    with torch.no_grad():
        test_text_matrix[idx] = xlnet_model(**inputs).last_hidden_state.squeeze(0).cpu().numpy()
        
    test_labels[idx] = int(row['label'])
    valid_indices.append(idx)

final_test_text = test_text_matrix[valid_indices]
final_test_image = test_image_matrix[valid_indices]
final_test_labels = test_labels[valid_indices]
final_test_labels_cat = to_categorical(final_test_labels, num_classes=2)

print(f"\n[INFO] Đã trích xuất thành công {len(valid_indices)} bài báo tập Test!")

# ==========================================
# 3. ĐÁNH GIÁ MÔ HÌNH
# ==========================================
loss, accuracy = model.evaluate(
    x=[final_test_text, final_test_image], 
    y=final_test_labels_cat,
    verbose=1
)

print("\n" + "="*50)
print(f"ĐIỂM TEST ACCURACY: {accuracy * 100:.2f}%")
print("="*50)
```
